```python
### 利用 selenium 爬蟲
import time 
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import openpyxl
```

```python
chrome_path = "C:/Users/hjkl/Desktop"
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument('window-size=2560,1440')
driver = webdriver.Chrome(options=options)

data = []
driver.get('https://ani.gamer.com.tw/animeList.php?sort=2')
driver.find_element(By.XPATH, '/html/body/dialog/form/div/div/div[2]/button[2]').click()

xpath1 = '/html/body/div[2]/div/div[1]/div[3]/a['
for i in range(10):
    i = i + 1
    
    h4 = driver.find_element(By.XPATH, xpath1 + str(i) + ']/div[2]/p')
    p2 = driver.find_element(By.XPATH, xpath1 + str(i) + ']/div[2]/div/p')
    p3 = driver.find_element(By.XPATH, xpath1 + str(i) + ']/div[2]/div/span')
    p4 = driver.find_element(By.XPATH, xpath1 + str(i) + ']/div[1]/div[2]/p')
    
    data.append({
        "作品名": h4.text,
        "年份": p2.text,
        "集數(截至今日)": p3.text,
        "觀看數": p4.text
    })

    time.sleep(3)

df = pd.DataFrame(data)
df.to_excel('Anime.xlsx', index=False)

driver.quit()
print('finish')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import urllib
import re
from matplotlib import font_manager as fm, rcParams

# 設置中文字體
font_path = 'C:/Users/hjkl/Desktop/新增資料夾/SimHei.ttf' 
font_prop = fm.FontProperties(fname=font_path)
rcParams['font.family'] = font_prop.get_name()

# 資料庫連接資訊
conn_string = (
    r'Driver={ODBC Driver 17 for SQL Server};'
    r'Server=DESKTOP-VUC3PO6\SQLEXPRESS01;'
    r'Database=2024年巴哈姆特動畫瘋月觀看前十名;'
    r'Trusted_Connection=yes;'
)

# 將連接字符串編碼為URL格式
conn_url = f"mssql+pyodbc:///?odbc_connect={urllib.parse.quote_plus(conn_string)}"

# 創建引擎
engine = create_engine(conn_url)
table_name = '7月'

# 從資料庫讀取資料
df = pd.read_sql(f'SELECT * FROM [{table_name}]', engine)

# 定義函數將字符串格式轉換為數字
def convert_to_number(s):
    match = re.match(r'(\d+\.?\d*)萬?', s)
    if not match:
        return 0
    number = float(match.group(1))
    return number * 1e4 if '萬' in s else number

# 應用轉換函數
df['觀看數'] = df['觀看數'].apply(convert_to_number)

# 取得前10筆數據
top_10 = df.head(10)
views = top_10['觀看數']
names = top_10['排名']

# 繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(views, labels=names, autopct='%1.1f%%', startangle=140)
plt.title('七月觀看次數圓餅圖', fontproperties=font_prop)
plt.tight_layout()
plt.savefig("./static/PieChart1.png")
plt.show()

# 繪製長條圖
plt.figure(figsize=(10, 6))
plt.bar(names, views)
plt.xlabel('排名', fontproperties=font_prop)
plt.ylabel('觀看次數(萬)', fontproperties=font_prop)
plt.title('七月觀看次數長條圖', fontproperties=font_prop)
plt.xticks(rotation=45, fontproperties=font_prop)
plt.tight_layout()
plt.savefig("./static/BarChart2.png")
plt.show()
```
